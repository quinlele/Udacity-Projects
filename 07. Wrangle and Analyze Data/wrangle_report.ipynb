# Wrangle Report
 
## 1. Introduction

The dataset that I was wrangling (and analyzing and visualizing) is the tweet archive of Twitter user @dog_rates, also known as WeRateDogs. WeRateDogs is a Twitter account that rates people's dogs with a humorous comment about the dog

This report briefly describes my work. The steps of wrangling this dataset are as follows:

• Gathering data

• Assessing data

• Cleaning data

## 2. Gathering data

My gathering data work gathered data from these sources

• Directly download the WeRateDogs Twitter archive data (twitter_archive_enhanced.csv), naming the dataset "twi_arch_enhanced"

• Use the Requests library to download the tweet image prediction (image_predictions.tsv), naming the dataset "img_pred"

• Use the Tweepy library to query additional data via the Twitter API (tweet_json.txt), naming the dataset "twe_json"
